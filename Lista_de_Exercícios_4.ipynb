```python

import random

palavras = ["RESTAURANTE", "ACABAXI", "CACHORRO", "TREM", "SUPERMERCADO", "GATO", "UNIVERSIDADE", 
"AEROPORTO", "TAPETE", "CORREDOR", "VARANDA", "JARDIM", "TRAVESSEIRO", "FUTEBOL", "CHURRASCO",
"BICICLETA", "BIBLIOTECA", "FLORESTA"]

#Palavra:
palavra = random.choice(palavras)

#Tamanho inicial das palavra:
formato = ["_"]*len(palavra)

#Lista composta pelas letras digitadas corretamente pelo jogador
letras_corretas = []

#Letras digitadas 
digitadas = []

#Número de tentativas
tentativas = 3

#Contador dos erros
erros = 0

#Contador dos acertos
acertos = 0


print("-"*40)
print("     Bem vindo ao Jogo da Forca")
print("-"*40)

print("\nInstruções:")
print("1- Tente adivinhar uma palavra.")
print("2- O jogo termina quando você errar 3 letras.")
print("3- A quantidade de letras na palavra está representada pelos traços abaixo.")

#Quantidade de letras

formato = ("_ ")*len(palavra)

print(f"Palavra: {formato} \n")


mensagem = ''

while True:
    letra = input("Digite uma letra: ").upper()
    print("\n")

    if len(letra) != 1:
            print ("Essa entrada não é válida. Apenas uma letra por vez é aceita. \n")
            continue
    
    if not letra.isalpha():
        print("Essa entrada não é válida. Apenas uma letra por vez é aceita.")
        continue

    digitadas += letra

    if digitadas.count(letra) > 1:
        print("Você já digitou essa letra. Tente outra letra. \n")
        continue   
    
    

    if letra in palavra:
        mensagem = ''
        letras_corretas += letra
        print("Parabéns. Você acertou uma letra. \n")
        for letra in palavra:
            if letra in letras_corretas:
                mensagem += letra
            else: 
                mensagem += "_ "
      
        if mensagem == palavra:
            print("Parabéns! Você ganhou o jogo.")
            print(f"A palavra secreta é: {mensagem}.")
            break

        else:
            print(f"Palavra: {mensagem} \n")
    
    else: 

        erros += 1
        
        if mensagem == '' and erros == 1:
        
            print("Você errou uma letra. Agora tem mais 2 chances.")
            print("  _______     ")
            print(" |/      |    ")
            print(" |      (_)   ")
            print(" |            ")
            print(" |            ")
            print("_|___        \n")

            print(f"A palavra é: {formato} \n")
        
        elif mensagem == '' and erros == 2:
            
            print("Você errou uma letra. Agora tem mais 1 chances.")
            print("  _______     ")
            print(" |/      |    ")
            print(" |      (_)   ")
            print(" |       |   ")
            print(" |       |    ")
            print("_|___         \n ")
            
            print(f"A palavra é: {formato} \n")
        

        elif erros == 1:
            
            print("Você errou uma letra. Agora tem mais 2 chances. ")
            print("  _______     ")
            print(" |/      |    ")
            print(" |      (_)   ")
            print(" |            ")
            print(" |            ")
            print("_|___        \n")
        
            print(f"A palavra é: {mensagem} \n")

        elif erros == 2:
            print("Você errou uma letra. Agora tem mais 1 chance. \n")

            print("  _______     ")
            print(" |/      |    ")
            print(" |      (_)   ")
            print(" |       |   ")
            print(" |       |    ")
            print("_|___        \n")

        
            print(f"A palavra é: {mensagem} \n")

        else:
            print("Você perdeu o jogo!")

            print("  _______     ")
            print(" |/      |    ")
            print(" |      (_)   ")
            print(" |      \|/   ")
            print(" |       |    ")
            print(" |      / \   ")
            print("_|___         ")

            print(f"A palavra secreta era: {palavra}.")
            break
```
